```python
missing_values = ["?"]
df = pd.read_csv('https://raw.githubusercontent.com/niteen11/DataAnalyticsAcademy/master/Python/dataset_diabetes/diabetic_data.csv',
na_values = missing_values)
df.head()
```

```python
drop_cols = ['admission_type_id', 'discharge_disposition_id', 'admission_source_id',
       'payer_code', 'medical_specialty',
       'number_outpatient', 'number_emergency', 'number_inpatient', 'diag_1',
       'diag_2', 'diag_3', 'number_diagnoses', 'max_glu_serum', 'A1Cresult',
       'metformin', 'repaglinide', 'nateglinide', 'chlorpropamide',
       'glimepiride', 'acetohexamide', 'glipizide', 'glyburide', 'tolbutamide',
       'pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
       'tolazamide', 'examide', 'citoglipton', 'insulin',
       'glyburide-metformin', 'glipizide-metformin',
       'glimepiride-pioglitazone', 'metformin-rosiglitazone',
       'metformin-pioglitazone', 'change', 'diabetesMed', 'readmitted']
```

```python
df.drop(drop_cols, inplace = True, axis=1)
df.columns
```

```python
df.isnull().sum()
```

```python
df = df.drop('weight', axis=1)
df.columns
```

```python
encounter_by_race = df.groupby('race')['encounter_id'].count().sort_values(ascending=False)
encounter_by_race
```

```python
df = df.astype({'encounter_id': 'str', 'patient_nbr': 'str', 'race':'str'})
df.dtypes
```

```python
df_Caucasian = df[df['race'].str.contains('Caucasian', na=False)]
df_Caucasian.describe()
```

```python
df_AfricanAmerican = df[df['race'].str.contains('AfricanAmerican', na=False)]
df_AfricanAmerican.describe()
```

```python
df_Hispanic = df[df['race'].str.contains('Hispanic', na=False)]
df_Hispanic.describe()
```

```python
df_Other = df[df['race'].str.contains('Other', na=False)]
df_Other.describe()
```

```python
df_Asian = df[df['race'].str.contains('Asian', na=False)]
df_Asian.describe()
```

```python
graph1 = df.groupby('race')['encounter_id'].count().sort_values(ascending=False).plot.bar(figsize = (10,5))
```
