```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv")
df.head()
```

```python
df.describe()
```

```python
desired_school1 = df[df['School Name'].str.contains('Newcomers')&(df['School DBN'].str.contains('Q'))]
desired_school1
```

```python
desired_school2 = df[df['School Name'].str.contains('Newtown')&(df['School DBN'].str.contains('Q'))]
desired_school2
```

```python
schools_merged = pd.concat([desired_school1, desired_school2], ignore_index=False)
schools_merged
```

```python
schools_merged.isnull().sum().sum()
```

```python
schools_merged.columns
```

```python
desired_columns = schools_merged[['School DBN', 'School Name', 'School Level', 'Regents Exam', 'Year','Total Tested', 'Mean Score', 'Number Scoring Below 65', 'Percent Scoring Below 65']]
desired_columns
```

```python
desired_regent = desired_columns[desired_columns['Regents Exam'].str.contains('Common Core Algebra2')]
desired_regent
```

```python
desired_regent.shape
```

```python
desired_regent['Number Scoring Below 65'] = pd.to_numeric(desired_regent['Number Scoring Below 65'], errors='coerce')
desired_regent
```

```python
group_by_year = desired_regent.groupby('Year')['Number Scoring Below 65'].count()
group_by_year
```

```python
grouping_sum = desired_regent.groupby(['School Name'])['Number Scoring Below 65'].sum()
grouping_sum
```

```python
total = desired_regent.groupby('School Name')['Total Tested'].sum()
total
```

```python
percentage_scoring_below_65 = (grouping_sum / total) * 100
percentage_scoring_below_65
```

```python
percentage_scoring_below_65.plot(kind = 'bar')
```

```python
#In conclusion, out of all students tested in the Common Core Algebra2 regent between 2016 and 2017, 35% of Newtown High School students' scored below 65 and 16% of Newcomers High School students' scored below 65.
#Newtown High School students' showed bigger deficiencies compared to Newcomers High School students'.
```
